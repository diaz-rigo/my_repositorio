# **Tratamiento de Valores Faltantes**

---



* Técnica Utilizada: Imputación por la moda (valor más frecuente).
* Técnica Utilizada: Imputación por la mediana.



```python
# Tratamiento de valores faltantes
for column in ['hear_left', 'hear_right', 'SMK_stat_type_cd']:
    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)

median_value = df['urine_protein'].median()
df['urine_protein'].fillna(median_value, inplace=True)
```

# **Escalado o normalización**

Dado que los atributos son de tipo ratio y pueden tener diferentes escalas, escalaremos los datos usando StandardScaler.


```python
# Escalado o normalización
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.drop('DRK_YN', axis=1))

# Crear un nuevo DataFrame con los datos escalados
scaled_df = pd.DataFrame(scaled_data, columns=df.columns[:-1])
scaled_df['DRK_YN'] = df['DRK_YN']
```

# **	Selección de atributos:  **

en este caso, todas las características podrían ser potencialmente relevantes para predecir la variable objetivo 'DRK_YN'.


```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

X = scaled_df.drop('DRK_YN', axis=1)
y = scaled_df['DRK_YN']

logreg = LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=1000)


selector = RFE(estimator=logreg, n_features_to_select=10, step=1)

selector = selector.fit(X, y)

print('Características seleccionadas:')
for i, feature in enumerate(X.columns):
    if selector.support_[i]:
        print(feature)
```

```python
X
```

```python
y
```

```python

X_optimal = X.loc[:, selector.support_]
```

# **Modelo SVM**


```python
# Modelo SVM
svm_model = SVC()
kf = KFold(n_splits=10, random_state=0, shuffle=True)
svm_scores = cross_val_score(svm_model, X_optimal, y, cv=kf, scoring='accuracy')
svm_accuracy = svm_scores.mean()
print(f'SVM Accuracy: {svm_accuracy}')
```

# **Modelo Random Forest**

```python
# Modelo Random Forest
rf_model = RandomForestClassifier()
rf_scores = cross_val_score(rf_model, X_optimal, y, cv=kf, scoring='accuracy')
rf_accuracy = rf_scores.mean()
print(f'Random Forest Accuracy: {rf_accuracy}')
```
